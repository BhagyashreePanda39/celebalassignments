```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\bhagy\Downloads\gas_prices.csv")

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

countries_to_plot = ['USA', 'Canada', 'Germany', 'UK', 'Italy']

plt.figure()
for country in countries_to_plot:
    plt.plot(df['Year'], df[country], marker='o', label=country)
plt.title('Gas Prices Over the Years (Selected Countries)')
plt.xlabel('Year')
plt.ylabel('Gas Price (USD per gallon)')
plt.legend()
plt.tight_layout()
plt.show()

year_data = df[df['Year'] == 2000].drop(columns='Year').T
year_data.columns = ['Price']
year_data = year_data.sort_values('Price', ascending=False)

plt.figure()
sns.barplot(x=year_data.index, y=year_data['Price'], palette="viridis")
plt.title('Gas Prices by Country in 2000')
plt.xlabel('Country')
plt.ylabel('Gas Price (USD per gallon)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
